9. What three functions can be used to get the integer, floating-point number, or string
version of a value?

````python
v = 25.6
print(f'type of variable : {type(v)}\t value : {v}')
print(f'type of variable wint int() function: {type(int(v))}\t value : {int(v)}')
print(f'type of variable wint float() function: {type(float(v))}\t value : {float(v)}')
print(f'type of variable wint str() function: {type(str(v))}\t value : {str(v)}')
````
