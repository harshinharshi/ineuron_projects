Write a Python program to do arithmetical operations addition and division.?

```python
def addition(a, b):
  '''function to do arithametic addition'''
  return(a+b)

def division(a,b):
  '''fuction to do arithameric division'''
  try :
    return(a/b)
  except Exception as e:
    return(e)

def main():
  while 1:
    v = int(input('enter 1 for addition or enter 2 for division or 0 for exiting'))
    if v == 0:
      print('exit')
      break
    a = int(input('enter first number'))
    b = int(input('enter second number'))
    if a<0 or b<0 :
      print('negative numbers not allowed')
      continue;
    if v == 1:
      print(addition(a, b))
    elif v == 2:
      print(division(a,b))
    else:
      print('invalid operation value')

if __name__ == '__main__':
  main()
```

Write a Python program to find the area of a triangle?

```python
def area_triangle(h,b):
  '''function to find the area of the triangle'''
  return(0.5*h*b)

def main():
  while 1:
    v = int(input('1-find area of triangel, 0- exit'))
    if v == 1:
      l = int(input('enter the height of triangle'))
      b = int(input('enter the base of triangle'))
      print(area_triangle(l,b))
    else:
      print('exit')
      break;
if __name__ == '__main__':
  main()
```

Write a Python program to swap two variables?

```python
a = 7
b = 5
print(f'a = {a}')
print(f'b = {b}')
tmp = a
a = b
b = tmp
print(f'\na = {a}')
print(f'b = {b}')
```
